```python
url='https://api.wazirx.com/sapi/v1/exchangeInfo'
```

```python
import requests
import json 
from types import SimpleNamespace
import pandas as pd
```

```python
df = requests.get(url)
```

```python
hello=json.loads(df.content,object_hook=lambda d : SimpleNamespace(**d))
hello
```

```python
symbol= []
for p in hello.symbols:
    symbol.append(p.symbol)
```

```python
status=[]
for i in hello.symbols:
    status.append(i.status)
```

```python
BaseAsset= []
for p in hello.symbols:
    BaseAsset.append(p.baseAsset)

```
